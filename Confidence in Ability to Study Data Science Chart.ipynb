```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.xkcd()

fig=plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.xticks([])
plt.yticks([])

ax.set_ylim([0,100])
ax.set_xlim([0,100])

sin_wave = np.sin(np.arange(25))*2

x_well = sin_wave+90 #Puts us at 30,90
x_sick = sin_wave+50 #np.ones(20)*5- np.arange(20)/1.4
x_sick_2 = np.ones(25)*10 #sin_wave*3.5+10 #np.ones(20)*5- np.arange(20)/1.4
#x_well_2

x = np.concatenate((x_well,x_sick,x_sick_2,x_well))

plt.annotate('I sneezed',xy=(25,87.5),arrowprops=dict(arrowstyle='->'),xytext=(40,90))

plt.annotate('I woke up feeling sick\nwhen I thought I would \nfeel well',xy=(49,50),arrowprops=dict(arrowstyle='->'),xytext=(30,70))

plt.annotate('I felt well \nenough to go \nfor a walk',xy=(73.5,11),arrowprops=dict(arrowstyle='->'),xytext=(58,40))

plt.plot(x)

plt.xlabel('time')
plt.ylabel('confidence');
plt.title('My confidence in my ability to study data science');
```
