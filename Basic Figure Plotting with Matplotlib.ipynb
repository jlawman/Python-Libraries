### Example: Simple plot

```python
x=np.linspace(0,5,10)
y=x **2

figure()  #Create an figure
plot(x,y,'r')    #Set the plot data of x,y with color red (try 'g' or 'b' for a different color)
xlabel('x')      #Set x label to read 'x'
ylabel('y')      #Set y label to read 'y'
title('Sample Plot')  #Set title
show()  #Show plot
```

### Example: Subplot with marker selection

Various markers will be used to display the data in the following plots. See https://matplotlib.org/api/markers_api.html for marker options

```python
#Subplot can be used to create multiple small plots in a grid
#The function takes the following args: subplot(nrows, ncols, plot_number)

nrows=2 #We will create a 2x2 grid of plots
ncols=2

subplot(nrows,ncols,1)
plot(x,y,'r--') #color red, dotted line

subplot(nrows,ncols,2)
plot(x,y,'b')  #color blue, default

subplot(nrows,ncols,3)
plot(x,y,'go')  #color green, marker circle

subplot(nrows,ncols,4)
plot(x,y,'g*-'); #color green, marker: star with line

#(note the ';' is not required, but makes a cleaner display)
```

### Example: Create specific fig for object-orient programming

```python
fig=plt.figure()
axes=fig.add_axes([.1,.1,0.8,0.8]) #Set dimensions ranges from (0,1)

axes.plot(x,y,'rv')    #plot data with red line, marker v

axes.set_xlabel('x')  #label axes
axes.set_ylabel('y')

axes.set_title('Sample plot');  #Set title
```

### Example: Multiple axis and figure within a figure

```python
fig=plt.figure()

axes1=fig.add_axes([.1,.1,.8,.8]) #main axis
axes2=fig.add_axes([.2,.5,.4,.3]) #inner figure axis

# main figure
axes1.plot(x,y,'rh')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Main figure')

# inner figure
axes2.plot(x,y,'gx')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('Inner figure');
```

### Example: Setting up plot with plt.subplots()

```python
fig, axes = plt.subplots()

axes.plot(x,y,'b+')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Sample plot');
```

### Example: Using subplots and for loop to create plots

```python
fig, axes = plt.subplots(nrows=1,ncols=3)
fig.tight_layout() #Not necessary, but it cleans the display

for ax in axes:
    ax.plot(x,y,'g--')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Sample Plot')
```

### Example: Creating grid of subplots

```python
fig, [row1, row2] = plt.subplots(nrows=2,ncols=4)
fig.tight_layout(h_pad=4)

for ax in row1:
    ax.plot(x,y,'g--')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Row 1 plot')
    
for ax in row2:
    ax.plot(x,y,'b--')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Row 2 plot')
```

### Example: Figure with specific figure-size

```python
fig=plt.figure(figsize=(8,2),dpi=100)#800x400 pixel 100 dpi image
axes=fig.add_axes([.1,.1,.8,.8])
```

### Example: Creating specific figure-size plot with subplot

```python
fig, axes= plt.subplots(figsize=(10,3))

axes.plot(x,y,'b>');
```
